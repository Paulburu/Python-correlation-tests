```python
#import libraries

import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None



# read in the data
df = pd.read_csv('movies.csv')
```

```python
#look at data
df.head()
```

```python
# check for missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
    
```

```python
#remove missing values

df = df.dropna()
```

```python
#data types
df.dtypes
```

```python
#adjust data type of columns

df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
```

```python
# create correct year column
#df['year_correct'] = df['released'].astype(str).str[:]
```

```python
#order data by gross
df.sort_values(by=['gross'], inplace=True, ascending=False)
df
```

```python
#check for duplicates

#df['name'].drop_duplicates().sort_values(ascending=False)
df.drop_duplicates()
```

```python
#Prediction: budget and company have high correlation with earnings
```

```python
#scatter plot with budget vs gross revenue

plt.scatter(x=df['budget'],y=df['gross'])

plt.title('Budget vs earnings')
plt.xlabel('budget')
plt.ylabel('earnings')
plt.show
```

```python
# plot budget vs gross using seaborn

sns.regplot(x='budget', y='gross', data=df, scatter_kws={'color':'red'}, line_kws={'color':'blue'})
```

```python
# look at correlation

df.corr()
```

```python
#visualize corr 

correlation_matrix = df.corr()

sns.heatmap(correlation_matrix, annot=True)
plt.title('Corr for numeric features')
plt.xlabel('Movie features')
plt.ylabel('Movie features')
plt.show()
```

```python
#replace data types to numbers

df_numerized = df

for col in df_numerized.columns:
    if(df_numerized[col].dtype == 'object'):
        df_numerized[col] = df_numerized[col].astype('category')
        df_numerized[col] = df_numerized[col].cat.codes
df_numerized
    
```

```python
#corr matrix again with numarized data

correlation_matrix = df_numerized.corr()

sns.heatmap(correlation_matrix, annot=True)
sns.set(rc = {'figure.figsize':(20,20)})
plt.title('Corr for numeric features')
plt.xlabel('Movie features')
plt.ylabel('Movie features')
plt.show()
```

```python
correlation_mat = df_numerized.corr()

corr_pairs = correlation_mat.unstack()

corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()
sorted_pairs
```
